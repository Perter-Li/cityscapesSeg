```python
import torch
from Yolo import YoloSeg
# 加载模型
# 这里将模型整个进行了封装，YoloSeg中存储了整个模型的结构，以及对应的训练器
model = YoloSeg('./yolov8n-seg.pt')  # load a pretrained model (recommended for training)
# 输出模型结构
print(model)
```

```python
# 进行模型训练
results = model.train(data='./cityscapes-seg.yaml', epochs=100, imgsz=640,batch=64)
```

```python
# 获取训练加载器
trainer = model.trainer
print(trainer)

# 查看训练数据格式
print(next(enumerate(trainer.train_loader)))
```

```python
# 查看优化器
optim = trainer.optimizer
print(optim)
```

```python
# 进行模型预测
from PIL import Image

# 选择一个预训练好的模型进行加载
model = YoloSeg('best.pt')  # load a custom model

results = model('processed_dataset/images/val/munster_000167_000019_leftImg8bit.png')  # predict on an image

# 显示结果
for r in results:
    im_array = r.plot(boxes=False)  # plot a BGR numpy array of predictions
    im = Image.fromarray(im_array[..., ::-1])  # RGB PIL image
    im.show()  # show image
    im.save('results.jpg')  # save image
```
